```python
#importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
#loading our dataset.
df = pd.read_csv("movies_metadata.csv")
df.head()
```

```python
# drop all duplicate rows.
df = df.drop_duplicates().reset_index(drop=True)
```

```python
print(df.duplicated().sum())
```

```python
# creating new dataframe with selected columns
df = df[['title','overview','genres','tagline','vote_average','popularity']]
```

```python
df.head()
```

```python
print(df.isna().sum())
```

```python
# drop rows where title is null.
df = df.dropna(subset=['title'])
```

```python
# filling with empty space.
df['overview'] = df['overview'].fillna('')
```

```python
df.iloc[0]['genres']
```

```python
import ast
# just only taking name from the df['genres'] dictionary.
df['genres'] = df['genres'].apply(lambda x : " ".join([i['name'] for i in ast.literal_eval(x)]))
```

```python
df.head(5)
```

```python
df['tagline'] = df['tagline'].fillna('')
```

```python
# merging all these three columns into tags.
df['tags'] = df['overview']+" "+df['genres']+" "+df['tagline']
```

```python
df.head(5)
```

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

```python
def preprocess_text(text):
    # converting to lowercase.
    text = str(text).lower()
    # next removing all the punctuation.
    text = re.sub(r'[^\w\s]', '', text)
    # next remove stop words.
    words = text.split()
    words = [word for word in words if word not in stop_words]
    # next lemmatizing the words.
    words = [lemmatizer.lemmatize(word) for word in words]
    return ' '.join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
df = df.reset_index(drop=True)
```

```python
# using indices for each title. using pd.Series
indices = pd.Series(df.index, index=df['title']).drop_duplicates()
indices.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
# converting text to vectors
# using tf-idf vectorizer.
tfidf = TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
# creating tfidf_matrix
tfidf_matrix = tfidf.fit_transform(df['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#using cosine similarity to get similarity scores to understand which movies are similar
def recommend(title, n=10):
  if title not in indices:
    return ['Movie not found']

  idx = indices[title]
  sim_scores = cosine_similarity(tfidf_matrix[idx], tfidf_matrix).flatten()
  similar_idx = sim_scores.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend('Avatar')
```

```python
recommend('Waiting to Exhale')
```

```python
recommend('The Godfather')
```
